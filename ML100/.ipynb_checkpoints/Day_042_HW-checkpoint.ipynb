```python
from sklearn import datasets, metrics

# 如果是分類問題，請使用 DecisionTreeClassifier，若為回歸問題，請使用 DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, export_graphviz
from sklearn.model_selection import train_test_split
from graphviz import Source
from IPython.display import display
from sklearn import tree
from IPython.display import SVG

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# 讀取 breast_cancer 資料集
breast_cancer = datasets.load_breast_cancer()
breast_cancer.data[:3]
```

```python

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, test_size=0.25, random_state=4)

# 建立模型
clf = DecisionTreeClassifier()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

```python
clf_feature_importances = list(zip(breast_cancer.feature_names, clf.feature_importances_))
clf_feature_importances_sorted = sorted(clf_feature_importances, key=lambda x: x[1], reverse=True)
print( "Feature importance: ", clf_feature_importances_sorted )
```

```python
import pydotplus

tree_export_graphviz = tree.export_graphviz(clf, out_file=None
   , feature_names=breast_cancer.feature_names, class_names=['0', '1', '2'] 
   , filled = True)
pydot_graph = pydotplus.graph_from_dot_data(tree_export_graphviz)
pydot_graph.write_png('output/042_original_tree.jpg')
# pydot_graph.set_size('"100,100"')
# pydot_graph.write_png('output/042_resized_tree.jpg')
image = mpimg.imread('output/042_original_tree.jpg')
plt.axis("off")
plt.figure(figsize=(100,100))
plt.imshow(image)
plt.show()
```

```python

# 建立模型
clf = DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=None,
 max_features=None, max_leaf_nodes=None, min_impurity_decrease=1e-07, min_samples_leaf=1,
 min_samples_split=3, min_weight_fraction_leaf=0.0, presort=False, random_state=None, splitter='best')

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)
```

# 2. 改用其他資料集 (boston, wine)，並與回歸模型的結果進行比較

```python
# 讀取 Boston 資料集
Wine = datasets.load_wine()
Wine.data[:3]
```

```python

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(Wine.data, Wine.target, test_size=0.25, random_state=4)

# 建立模型
clf = DecisionTreeClassifier()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

```python
clf_feature_importances = list(zip(breast_cancer.feature_names, clf.feature_importances_))
clf_feature_importances_sorted = sorted(clf_feature_importances, key=lambda x: x[1], reverse=True)
print( "Feature importance: ", clf_feature_importances_sorted )
```

```python

tree_export_graphviz = tree.export_graphviz(clf, out_file=None
   , feature_names=Wine.feature_names, class_names=['0', '1', '2'] 
   , filled = True)
pydot_graph = pydotplus.graph_from_dot_data(tree_export_graphviz)
pydot_graph.write_png('output/042_2_original_tree.jpg')
image = mpimg.imread('output/042_2_original_tree.jpg')
plt.axis("off")
plt.figure(figsize=(100,100))
plt.imshow(image)
plt.show()
```
